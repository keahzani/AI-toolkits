```python
# Task 1: Classical ML with Scikit-learn (Iris Dataset)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
import pandas as pd
import numpy as np

print("--- Task 1: Decision Tree Classifier ---")

# 1. Data Loading and Initial Inspection
iris = load_iris(as_frame=True)
df = iris.frame
X = iris.data
y = iris.target
target_names = iris.target_names

# The Iris dataset is usually clean, but we check for common preprocessing needs.
print(f"Dataset shape: {df.shape}")
print(f"Missing values count: {df.isnull().sum().sum()}") # Check for missing values

# 2. Preprocess Data (No missing values in Iris, but we'll show label encoding for context)
# The target 'y' is already label-encoded (0, 1, 2), so no further encoding is needed for the target.
# If features were categorical, we would use OneHotEncoder.

# 3. Split Data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 4. Train Model (Decision Tree)
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
print("\nDecision Tree Model Trained Successfully.")

# 5. Evaluation
y_pred = dt_classifier.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
# Use average='macro' for multi-class classification to average metrics across all classes
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print("\n--- Evaluation Metrics ---")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (Macro): {precision:.4f}")
print(f"Recall (Macro): {recall:.4f}")

# Example: Display a sample prediction
sample_index = 0
print(f"\nSample Prediction:")
print(f"Features: {X_test.iloc[sample_index].values}")
print(f"Actual Species: {target_names[y_test.iloc[sample_index]]}")
print(f"Predicted Species: {target_names[y_pred[sample_index]]}")
```

```python
# Task 2: Deep Learning with TensorFlow (MNIST CNN)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
import numpy as np

print("\n--- Task 2: CNN for MNIST Digits ---")

# 1. Data Loading
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# 2. Preprocessing
# Normalize images to a 0-1 range
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape data to include a channel dimension (28x28x1 for grayscale)
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)

# Convert labels to one-hot encoding (not strictly needed for sparse_categorical_crossentropy)
# y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
# y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)
input_shape = X_train.shape[1:]

# 3. Build CNN Model
model = Sequential([
    # Convolutional Layer: Learns features from the image (edges, corners)
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    # Pooling Layer: Reduces dimensionality, helping model generalize
    MaxPooling2D((2, 2)),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    
    # Flatten: Converts 2D feature maps into a 1D vector for the Dense layers
    Flatten(),
    
    # Dense Layers (Fully Connected): Performs classification
    Dense(100, activation='relu'),
    Dense(10, activation='softmax') # Output layer (10 classes: 0-9)
])

# 4. Compile and Train Model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # Use if labels are integers (0-9)
              metrics=['accuracy'])

print("\nCNN Model Architecture:")
model.summary()

# Training (Run this in Colab with GPU for speed)
history = model.fit(X_train, y_train, epochs=5, 
                    validation_data=(X_test, y_test), verbose=1)

# 5. Evaluation (Goal: >95% accuracy)
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {accuracy:.4f} (Goal: >0.95)")

# 6. Visualize Predictions on 5 Samples
predictions = model.predict(X_test)
fig, axes = plt.subplots(1, 5, figsize=(12, 3))
axes = axes.flatten()

for i in range(5):
    # Get image and prediction
    img = X_test[i].reshape(28, 28)
    predicted_digit = np.argmax(predictions[i])
    true_digit = y_test[i]
    
    axes[i].imshow(img, cmap='gray')
    axes[i].set_title(f"True: {true_digit}\nPred: {predicted_digit}", 
                      color="green" if predicted_digit == true_digit else "red")
    axes[i].axis('off')

plt.tight_layout()
plt.show() #
```

```python
# Task 3: NLP with spaCy (NER and Sentiment Analysis)
import spacy

# Load a small pre-trained English model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading spaCy model. Run the following command in your terminal/notebook: python -m spacy download en_core_web_sm")
    raise

print("\n--- Task 3: spaCy for NER and Sentiment ---")

# Sample Amazon Product Reviews (Text Data)
reviews = [
    "The new Sony Alpha a7 IV camera is fantastic, but the battery life is surprisingly short.",
    "I love the sound quality of these Bose QuietComfort Earbuds II, they block out everything!",
    "This budget laptop from Acer is slow, the plastic chassis feels cheap, and the charger broke quickly."
]

print("\n--- Named Entity Recognition (NER) ---")
for i, review in enumerate(reviews):
    doc = nlp(review)
    print(f"\nReview {i+1}: {review}")
    
    entities = []
    # Extract entities focused on products, brands, and organizations
    for ent in doc.ents:
        if ent.label_ in ['PRODUCT', 'ORG', 'NORP']:
            entities.append(f"'{ent.text}' ({ent.label_})")
            
    print(f"Extracted Entities: {', '.join(entities) if entities else 'None found'}")


print("\n--- Rule-Based Sentiment Analysis ---")

# Rule-Based Sentiment Analysis: Count positive/negative keywords
positive_words = ["fantastic", "love", "great", "excellent", "block out"]
negative_words = ["short", "slow", "cheap", "broke", "disappointing"]

def analyze_sentiment(text):
    doc = nlp(text.lower())
    pos_count = sum(1 for token in doc if token.text in positive_words)
    neg_count = sum(1 for token in doc if token.text in negative_words)
    
    if pos_count > neg_count:
        return "Positive"
    elif neg_count > pos_count:
        return "Negative"
    else:
        return "Neutral"

for i, review in enumerate(reviews):
    sentiment = analyze_sentiment(review)
    print(f"Review {i+1}: Sentiment is {sentiment}")
```
